```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Importanje svih potrebnih paketa
```

```python
df = pd.read_csv('wdbc.data',header=None)
#Ovdje sam stavio 'header = None ' jer mi je bez njega prvi red stavljao kao column names
```

```python
df
```

```python
df.columns = ['ID', 'Diagnosis','radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean',
             'concavity_mean','concave_points_mean','symmetry_mean','fractal_dimension_mean','radius_sd','texture_sd',
              'perimeter_sd','area_sd','smoothness_sd','compactness_sd','concavity_sd','concave_points_sd',
              'symmetry_sd','fractal_dimension_sd','radius_mw','texture_mw','perimeter_mw','area_mw','smoothness_mw',
              'compactness_mw','concavity_mw','concave_points_mw','symmetry_mw','fractal_dimension_mw']

#Radi lakšeg snalaženja preimenovao sam kolone u nazive koji su u radu
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
le = LabelEncoder()
le.fit(df.loc[:,'Diagnosis'])
df.loc[:,'Diagnosis'] = le.transform(df.loc[:,'Diagnosis'])
df
#Kao što možemo vidjeti u gornjim tablicama Diagnosis je imao oznake M = malignant, B = benign, importao sam LabelEncoder i 
# pretvorio M = 1 i B = 1 radi lakše analize podataka
```

```python
sns.countplot(x = 'Diagnosis', hue = 'Diagnosis', data = df , dodge=False)
plt.yticks([0, 100, 200, 300])
plt.show()
#Predstavio sam identičan countplot kao i u radu koji nam pokazuje omjer malignant(1) i benign(0) tumora
```

```python
n_df = df[['Diagnosis','radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean',
             'concavity_mean','concave_points_mean','symmetry_mean','fractal_dimension_mean']]
n_df
#Priprema df-a za heatmap 
```

```python
correlation_matrix = n_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```
